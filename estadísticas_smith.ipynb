```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url_ozzie = "https://www.baseball-reference.com/players/s/smithoz01.shtml"
response = requests.get(url_ozzie)
html_content = response.content

soup = BeautifulSoup(html_content, "html.parser")

tabla_bateo_ozzie = soup.find("table", {"class": "stats_table"})
if tabla_bateo_ozzie:
    datos_ozzie = []
    encabezados_ozzie = [header.get_text() for header in tabla_bateo_ozzie.find_all("th")]
    encabezados_ozzie = encabezados_ozzie[1:]  

    filas_ozzie = tabla_bateo_ozzie.find_all("tr")
    for fila_ozzie in filas_ozzie[1:]:
        fila_datos_ozzie = {}
        celdas_ozzie = fila_ozzie.find_all(["th", "td"])
        for idx, celda_ozzie in enumerate(celdas_ozzie):
            fila_datos_ozzie[encabezados_ozzie[idx - 1]] = celda_ozzie.get_text() 
        datos_ozzie.append(fila_datos_ozzie)

    df_ozzie = pd.DataFrame(datos_ozzie)
    print(df_ozzie)
else:
    print("No se encontró la tabla de estadísticas de bateo para Ozzie Smith.")
```
